```python
def plot_pie_charts(df, title):
    fig = plt.figure(figsize=(15, 5), dpi=80)
    titles = ["Nula", "Baixa", "Alta"]
    fig.suptitle(title, fontsize="16")
    fig.subplots_adjust(hspace=0.4, wspace=0.4)
    for i in range(1, 4):
        ax = fig.add_subplot(1, 3, i)
        values = []
        labels = []
        for f_index in df.index.get_level_values(0).unique(): 
            values.append(df.loc[f_index, str(i-1)])
            labels.append(f_index)
        ax.pie(values, labels=labels, autopct='%1.1f%%',
            shadow=True, startangle=90)
        ax.set_title(titles[i-1])
        
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, 10),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

def plot_total_bar_chart(df, y_label, title):
    fig, ax = plt.subplots(figsize=(5, 5), dpi=80)
    barWidth = 0.30
    values = []
    labels = []
    for f_index in df.index.get_level_values(0).unique(): 
        values.append(df[f_index].sum())
        labels.append(f_index)

    r = 1
    for i in range(1, len(values)+1):
        r = r+barWidth
        bar_chart = ax.bar(r, values[i-1], width=barWidth, edgecolor='white', label=labels[i-1])
        #autolabel(bar_chart)
    
    ax.set_ylabel('Nº')
    ax.set_title(title)
    ax.set_xticks([barWidth])
    ax.legend()

    fig.tight_layout()

    plt.show()
    
def plot_columns_bar_chart(df, width_p, title):
    
    sum_width = 0
    labels = ["Nula", "Baixa", "Alta"]
    labels_bar = df.index.get_level_values(0).unique()
    values_list = []
    for f_index in df.index.get_level_values(0).unique(): 
        values = []
        for i in range(1, 4):
            values.append(np.round((df.loc[f_index, str(i-1)]/df[f_index].sum())*100,2))
        values_list.append(values)

    X = np.arange(3)
    fig = plt.figure(figsize=(10, 5), dpi=80)
    ax = fig.add_axes([0,0,1,1])
    for i in range(1, len(labels_bar)+1):
        bar_chart = ax.bar(X + sum_width, values_list[i-1], width = width_p, label=labels_bar[i-1])
        sum_width = sum_width + width_p
        #autolabel(bar_chart)

    ax.set_ylabel('%')
    ax.set_title(title)
    ax.set_xticks(X+width_p/2)
    ax.set_xticklabels(labels)
    ax.legend()
```

```python
import scipy.stats as ss
def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x,y)
    chi2 = ss.chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2/n
    r,k = confusion_matrix.shape
    phi2corr = max(0, phi2-((k-1)*(r-1))/(n-1))
    rcorr = r-((r-1)**2)/(n-1)
    kcorr = k-((k-1)**2)/(n-1)
    return np.sqrt(phi2corr/min((kcorr-1),(rcorr-1)))
```
